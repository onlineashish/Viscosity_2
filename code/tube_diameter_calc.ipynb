```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Read the image
image = cv2.imread("/home/cvig13/Viscosity_2/data/frames_first/frame0.jpg")  # Replace 'your_image.jpg' with your image path

# Get the dimensions of the image
height, width = image.shape[:2]

# Calculate the cropping margins (30% from left and 30% from right)
crop_left = int(width * 0.4)
crop_right = int(width * 0.7)

# Crop the image
cropped_image = image[:, crop_left:crop_right]




# Convert the image to grayscale
image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
# Display the image using Matplotlib
plt.imshow(image, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()

# edges = cv2.Canny(image, 20, 40)
# plt.imshow(edges, cmap='gray')  
# plt.axis('off')
# plt.show()

# # Perform Hough Line Transform to detect lines (adjust parameters for better line detection)
# lines = cv2.HoughLinesP(edges, 1, np.pi / 180, threshold=100, minLineLength=100, maxLineGap=10)
# # Draw detected lines
# line_image = np.zeros_like(image)
# for line in lines:
#     x1, y1, x2, y2 = line[0]
#     cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 3)
# # Show result
# plt.imshow(line_image, cmap='gray')
# plt.axis('off')
# plt.show()
```
